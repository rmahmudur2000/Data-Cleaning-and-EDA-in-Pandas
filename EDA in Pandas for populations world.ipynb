```python
# Selecting relevant columns for correlation calculation
numeric_columns = ['Rank',
    '2022 Population', '2020 Population', '2015 Population',
    '2010 Population', '2000 Population', '1990 Population',
    '1980 Population', '1970 Population', 'Area (km²)',
    'Density (per km²)', 'Growth Rate', 'World Population Percentage'
]

# Compute the correlation matrix
correlation_matrix = df[numeric_columns].corr()

# Round the correlation matrix to 2 decimal places
correlation_matrix_rounded = correlation_matrix.round(2)

# Display the rounded correlation matrix
correlation_matrix_rounded
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix_rounded, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Select numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Group by Continent and calculate the mean for all numerical columns
df2 = df.groupby('Continent')[numerical_columns].mean()

# Sort by mean 2022 Population in descending order
df2 = df2.sort_values('2022 Population', ascending=False)

# Display the grouped data
df2
```
